```python
bigbox= flipcart_html.find_all('div', {"class":"_1AtVbE col-12-12"})
```

```python
# To know for how many products, we are able to get bigboxes:
len(bigbox)
```

```python
# Since the first 3 bigboxes are header detail, We delete the first 3 boxes:
del bigbox[0:3]
```

```python
# To get to a particular page:
product_link = "https://www.flipkart.com/search?q=" + bigbox[0].div.div.div.a["href"]
```

```python
# We use for loop to iterate through all pages:
for i in bigbox:
    print("https://www.flipkart.com/search?q=" + i.div.div.div.a["href"])
```

```python
product_link
```

```python
# We scrap everything inside this particular product (product_link):
product_req = requests.get(product_link)
```

```python
product_req
```

```python
# To readout:
product_req.text
```

```python
# To transfer it into a human readable format:
product_html = bs(product_req.text, "html.parser")
```

```python
product_html
```

```python
comment_box = product_html.find_all('div', {"class":"_16PBlm"})
```

```python
# To check how many comments we have:
len(comment_box)
```

```python
# To get the name of the reviewer:
comment_box[0].div.div.find_all('p', {"class":"_2sc7ZR _2V5EHH"})[0].text
```

```python
# To get all the reviewers' names, we use the for loop to iterate through each comment:
for i in comment_box:
    print(i.div.div.find_all('p', {"class":"_2sc7ZR _2V5EHH"})[0].text)
```

```python
# To get the ratings from a particular reviewer:
comment_box[0].div.div.div.div.text
```

```python
# To get all the ratings:
for i in comment_box:
    print(i.div.div.div.div.text)
```
